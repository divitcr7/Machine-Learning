```python
df = pd.read_csv("who_is_responsible_for_global_warming.csv")
df.head()
```

```python
plt.figure(figsize=(13,9))
sns.heatmap(df)
```

- black color is minimum value
- Bright color is maximum value
- Eg. usa co2 eimssion was high then became low
- Eg 2 india Co2 emmision was low forever

```python
# change heatmap color ?
plt.figure(figsize=(13,9))
sns.heatmap(df, cmap = 'gnuplot_r', annot = True)
```

```python
# HeatMap label, title, fontrsize

plt.figure(figsize=(19,11))
ax = sns.heatmap(df, annot = False)
ax.set(title= 'HeatMap of "Countries Responsible For Global Warming',
      xlabel = 'Co2 Emission In every Country per year',
      ylabel = 'Names of the countries responsible',)
sns.set(font_scale = 2)
```

Correlation?
- stats mesaure of relation b/w 2 variables,
- values exist b/w 0 -> 1
- a. -1 means perfect negative correlation ( x decrease then y decrease )
    ( Eg. car distance inc , the petrol dec ).
  * b.  0 No correlation ( x inc or dec y will have no affect )
  Eg. ( car traveling but petrol in our home wont be affected )
  * c +1 Perfect positive correlation ( if X inc then Y will also inc )
  Eg. ( if your siblings are studying then you should also study yoy get motivated )

```python
# pearson bydefault
df.corr()
```

```python
# need to visualize heatmap properly
plt.figure(figsize = (16,9))
ax = sns.heatmap(df.corr(), annot = True, linewidth = 6
                 
            , annot_kws={"fontsize":11},xticklabels=1, yticklabels=1)

ax.set(title= 'HeatMap of "Countries Responsible For Global Warming',
      xlabel = 'Co2 Emission In every Country per year',
      ylabel = 'Names of the countries responsible',)
sns.set(font_scale = 0.4)
```

- Here we can say that there is strong positive correlation example in the bottom graph we can clearly see that the corr b/w 2014 & 2000 is very high and it keeps on increasing when compared with other years eg in bottom,
- 2014 - 2001, 2014 - 2002 .... 2014 - 2013 the corr is positive in nature as x incereases y also increases this concludes that co2 emmision is higher

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
# data like dictionaory but data to dataframe
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']],
                               columns = np.append(cancer_dataset['feature_names'], ['target']))
cancer_df
```

-  we can use all the 30 features & train the model but what if some of the features do not correlate with the target feature this might cause harm to the accuracy.

```python
sns.heatmap(cancer_df)
```

```python
plt.figure(figsize = (45, 30 ))
ax = sns.heatmap(cancer_df.corr(), annot = True, linewidth = 3, annot_kws={"fontsize":21},xticklabels=1, yticklabels=1)
ax.set(label = 'Breast cancer correlation',
      xlabel = 'Figure relation measurements', 
      ylabel = 'values')
sns.set(font_scale = 20)
```
