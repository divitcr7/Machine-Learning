```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.read_csv('D:\\Data Science\\py-master\\ML\\1_linear_reg\\Exercise\\canada_per_capita_income.csv')
```

```python
df.head()
```

```python
# plotting a scatter plot

plt.scatter(df[['year']], df[['per capita income']], color = 'blue', marker = 'x')
plt.xlabel('Years')
plt.ylabel('Income')
```

```python
# plotting a normal line 
plt.plot(df[['year']], df[['per capita income']])
```

```python
reg =linear_model.LinearRegression()

# train the model by fittinh
reg.fit(df[['year']], df[['per capita income']])
```

> As The model is trained it's ready for prediction

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.scatter(df[['year']], df[['per capita income']], color = 'blue', marker = 'x')
plt.xlabel('Years')
plt.ylabel('Income')
plt.plot(df[['year']],reg.predict(df[['year']]))
```

```python
# performing The same regression plot useing seaborn

sns.regplot(df[['year']],df[['per capita income']], data = df,marker = 'x')
```
