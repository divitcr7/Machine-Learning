```python
import pandas as pd

df = pd.read_csv('C:\\Users\\divit\Desktop\\Linear Regression ( Program + Use case )\\homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df[['town']])
dummies
```

```python
# merging the normal data frame & dummies data frame    
# or axis = 'columns'
merge = pd.concat([df,dummies], axis = 1)
merge.head()
```

```python
# remove the original town column & town_west windsor to get rid of dummy variable trap
final = merge.drop(['town','town_west windsor'], axis = 1)
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# seprating dependent & independent variable
x = final.drop(['price'], axis = 1)
```

```python
# we removed our dependent variable price & got out or independent variables
x.head()
```

```python
# This is our dependent variable
y = final[['price']]
y.head()
```

```python
# Training the Ml Model

model.fit(x,y)
```

```python
# predicting price of 2800sq ft home in robinsville

model.predict([[2800,0,1]])
```

```python
# predicting price of 2800sq ft home in windsor

model.predict([[2800,0,0]])
```

```python
# how accurate my model is 
model.score(x,y)
```
